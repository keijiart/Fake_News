```python
import pandas as pd
import numpy as np
```

```python
news = pd.read_csv('news_y.csv')
fake = pd.read_csv('Fake_y.csv')
true = pd.read_csv('True_y.csv')
```

```python
news.head()
```

```python
news['label'] = news['label'].map({'FAKE':0, 'REAL':1}).astype(int)
news.head()
```

```python
news.rename(columns={'label': 'class'}, inplace=True)
news.head()
```

```python
news = news.drop(['Unnamed: 0'], axis=1)
news.head()
```

```python
fake_news = news[news['class'] == 0]
fake_news
```

```python
true_news = news[news['class'] == 1]
true_news
```

```python
fake_marge = pd.concat([fake_news, fake], axis=0)
fake_marge = fake_marge.sample(frac = 1)
fake_marge.reset_index(inplace = True)
fake_marge.drop(["index"], axis = 1, inplace = True)
fake_marge['class'] = 0
fake_marge.drop(['class'], axis=1, inplace=True)
fake_marge
```

```python
true_marge = pd.concat([true_news, true], axis=0)
true_marge = true_marge.sample(frac = 1)
true_marge.reset_index(inplace = True)
true_marge.drop(["index"], axis = 1, inplace = True)
true_marge['class'] = 1
true_marge.drop(['class'], axis=1, inplace=True)
true_marge
```

```python
fake_marge.drop(fake_marge.index[-50:-1])
```

```python
true_marge.drop(true_marge.index[-50:-1])
```

```python
fake_marge.to_csv('Fake_estesi.csv', index=False)
```

```python
true_marge.to_csv('True_estesi.csv', index=False)
```

```python
fake_tail = fake_marge.tail(50)
```

```python
true_tail = true_marge.tail(50)
```

```python
fake_true = pd.concat([true_tail, fake_tail], axis=0)
fake_true
```

```python
fake_true = fake_true.sample(frac = 1)
fake_true.reset_index(inplace = True)
fake_true.drop(["index"], axis = 1, inplace = True)
fake_true.head()
```
